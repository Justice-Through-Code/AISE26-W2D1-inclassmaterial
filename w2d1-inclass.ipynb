#### Converting JSON entries to a table
Unzip the aircraft zipfile, read in each JSON entry, and create a single comma-delimited .csv file called 'aircraft.csv' with the columns "manufacturer", "model", "introduced", "length_ft", "top_speed_mph", and "number_of_engines".

````python
import os
import json
import csv

FOLDER = r'C:\Users\sstucker\Documents\JTC\AISE 25\files\aircraft'

with open('aircraft.csv', 'w') as wf:  # Open a CSV file for writing
    header = ["manufacturer", "model", "introduced", "length_ft", "top_speed_mph", "number_of_engines"]
    writer = csv.DictWriter(wf, fieldnames=header)  # Defines the names of the columns
    for file in os.listdir(FOLDER):  # Iterating through all the JSON files in the folder
        if file.endswith('.json'):
            with open(os.path.join(FOLDER, file), 'r') as f:  # Open them
                try:
                    d = json.load(f)
                except json.JSONDecodeError as e:
                    print(f"Error decoding JSON from {file}: {e}")
                    continue
                row = {key : str(d[key]) for key in header}
                # row = [d[key] for key in header]  # works if `writer` is csv.writer
                writer.writerow(row)  # Write only the values, not the keys

import pandas as pd
df = pd.read_csv('aircraft.csv')
df
````

````python
import json
import urllib

CITY = 'Brooklyn'

PATH_TO_KEY = r'C:\Users\sstucker\Documents\JTC\AISE 25\api.env'
with open(PATH_TO_KEY, 'r') as f:
    api_key = f.read().strip().split('=')[1]

params = {
    'q': CITY,
    'appid': api_key,
    'units': 'imperial',
    'limit': 1
}

url = "http://api.openweathermap.org/data/2.5/weather?" + urllib.parse.urlencode(params)

response = urllib.request.urlopen(url)

d = json.loads(response.read())

# print(json.dumps(d, indent=2))

feels_like = d['main']['feels_like']
print('Feels like', float(feels_like), 'degrees in', CITY)
````
